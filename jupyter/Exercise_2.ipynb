Q: *Can you explain what numbers are included in the 'others' array?* \
A: all numbers which are not divisible by 6

Q: *How can we simply replace the third and forth 'if' statements with a single 'if' and 'else' condition?* \
A: *see below*

```python
# Creating an array with values from 1 to 20
numbers = [i for i in range(1, 21)]

three_mult = []
two_mult = []
six_mult = []
others = []

for num in numbers:
    if num % 2 == 0:
        two_mult.append(num)
    if num % 3 == 0:
        three_mult.append(num)
    if num % 3 == 0 and num % 2 == 0:
        six_mult.append(num)
    #if num % 3 != 0 or num % 2 != 0:  # line could be replaced with:
    else:
        others.append(num)
    # Replacing with %6 + if and else


print(three_mult, two_mult, six_mult, others)
```

# Object Oriented Programming (OOP)
You can create custom classes, including inherited classes.

They can have fields and methods. Inside the class definition, the object is called `self`. `self` has to be the first argument of all normal (non-static) methods.

```python
class Foo:
    # constructor with one argument
    def __init__(self, some_arg):
        # add a field
        self.x = some_arg

        # private fields are not supported.
        # fields that are not supposed to be public are prefixed with a _:
        self._private = {}

    # a method that acts on an instance of this class
    def some_method(self):
        return self.x

# one or multiple base classes can be specified
class Derived(Foo):
    def reset_x(self):
        self.x = None

# create an instance of the class:
instance = Derived("e")

# method call syntax:
print(instance.some_method())
instance.reset_x()
print(instance.x)

# this would violate encapsulation:
print(instance._private)
```

Objects are passed by reference. Memory management is done automatically by language mechanisms (garbage collection).

Fields and methods are looked up at runtime, allowing for polymorphism without any inherent compile-time checks.

OOP is not mandatory in Python (unlike Java/C#). Many features are supported nonetheless:

## Getters, setters and operator overloading
Operators that can be overloaded with the `__underscore__` syntax include: arithmetic and comparison operators, conversions to primitive types, behavior of global functions like `len(...)` or `abs(...)`, indexing, iteration, function call syntax

Even normal attribute access syntax (the dot operator: `my_obj.field`) can be overridden. The system is fairly complex.

```python
class Wrapper:
    abc = 3

    def __init__(self, x: int) -> None:
        self._num = x

    @property
    def num(self) -> int:
        return self._num

    def __add__(self, other: Wrapper) -> Wrapper:
        return Wrapper((self._num + other._num) % 16)

a, b = Wrapper(9), Wrapper(11)
print((a + b).num)
```

## More advanced types
Python also supports abstract base classes. They can be used to represent some interface or trait which is available on the derived classes.

Enums are implemented entirely using library features, which is unusual.

```python
from abc import ABCMeta, abstractmethod
from enum import Enum, auto

class MyABC(metaclass=ABCMeta):
    @abstractmethod
    def override_me(self) -> None:
        ...

# cannot instantiate directly: MyABC()

class DerivedFromABC(MyABC):
    def override_me(self) -> None:
        print("test")

DerivedFromABC().override_me()

class Color(Enum):
    # automatically counts up
    RED = 1
    GREEN = auto()
    BLUE = auto()

Color.RED
```

## Matplotlib
Most famous Python library for visualization and plots. We would specifically use matplotlib a lot in future for plotting our audio samples in different domains. 

- *Take a look at this simple plot and try to understand the variables that are used for plotting. Optionally you can change them to your liking or lookup all of the available options to change the appearance, size, font, fontsize, line thickness, etc. of this plot*

```python
# Defining the time variable between 0 and 10 with 1 step increments
t = np.arange(start=0, stop=10, step=1)
# Generating random integer values between -2 and 2 with the exact same shape (number of elemnts) as our time variable
rng = np.random.default_rng()
F = rng.integers(low=-2, high=2, size=t.shape)

# Create a line plot
plt.plot(t, F, color='red', marker='o', linestyle='--', label='Random Plot')

# Add title and labels
plt.title("Simple Line Plot")
# X and Y axis labels
plt.xlabel("t (seconds)")
plt.ylabel("F")
# Adding the grid to our plot for easier readibility
plt.grid()
plt.legend()
```
